2. Teste de Hipoteses

```python
df_loop_aero = pd.read_csv('/datasets/project_sql_result_07.csv')
```

#Ele contém dados sobre viagens do Loop para o Aeroporto Internacional O'Hare. Lembre-se, estes são os valores dos campos da tabela:
start_ts:data e hora de coleta
weather_conditions:condições meteorológicas no momento em que a corrida começou
duration_seconds:duração da corrida em segundos

```python
df_loop_aero
```

```python
print(df_loop_aero.info())
```

```python
# Criar o boxplot
plt.figure(figsize=(8,6))
sns.boxplot(data= df_loop_aero, x='weather_conditions', y='duration_seconds')

# Personalizar o gráfico
plt.title('Comparação da duração das corridas\nentre clima Bom e Ruim (Sábados)')
plt.xlabel('Condições Climáticas')
plt.ylabel('Duração da Corrida (segundos)')
plt.grid(True)
plt.show()
```

# Teste a hipótese:

"A duração média dos passeios do Loop para o Aeroporto Internacional O'Hare muda nos sábados chuvosos." 
Esta nova tabela contem apenas dados de Sabados.
Hipótese:
H0 (nula): A duração média das corridas não muda nos sábados chuvosos. (nao muda em nada o tempo Bad ou Good)
H1 (alternativa): A duração média das corridas muda nos sábados chuvosos.(tempo pode alterar a media das corridas)

```python
# Verificar dados
print(df_loop_aero['weather_conditions'].value_counts())

# Separar as durações por tipo de clima
bad_weather = df_loop_aero[df_loop_aero['weather_conditions'] == 'Bad']['duration_seconds']
good_weather = df_loop_aero[df_loop_aero['weather_conditions'] == 'Good']['duration_seconds']

# Definir o nível de significância
alpha = 0.05

# Aplicar o teste t de Student para amostras independentes (assumindo variâncias diferentes)
results = stats.ttest_ind(good_weather, bad_weather, equal_var=False)

# Exibir resultados
print('Valor-p:', results.pvalue)

if results.pvalue < alpha:
    print('Rejeitamos a hipótese nula: Existe diferença na duração das corridas entre tempo bom e ruim.')
else:
    print('Não rejeitamos a hipótese nula: Não encontramos diferença significativa.')
```

Como formulamos as hipóteses?
Hipótese Nula (H0): A duração média das corridas é a mesma entre tempo "Good" e "Bad".
(Não há diferença significativa.)

Hipótese Alternativa (H1): A duração média das corridas muda entre tempo "Good" e "Bad".
(Há diferença significativa.)

Qual o critério para tomar decisão?
Definimos um nível de significância (α), geralmente 0,05 (ou seja, 5% de chance de erro).

Rodamos o t-test e calculamos o p-valor.

Se p-valor < alfa, rejeitamos H0 → existe diferença significativa.

Usei o t-test porque:
Comparo duas médias.
As amostras são independentes.
Os dados são contínuos.
Temos uma quantidade razoável de dados para confiar no resultado.
